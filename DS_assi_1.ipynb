```python
from sklearn.preprocessing import MinMaxScaler
```

```python
encode=MinMaxScaler()
```

```python
normalized=encode.fit_transform(df[['sepal_length','sepal_width','petal_length','petal_width']])
```

```python
normalized
```

```python
columns=['Normalized_sepal_length','Normalized_sepal_width','Normalized_petal_length','Normalized_petal_width']
```

```python
pd.DataFrame(normalized,columns=columns)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode_ohe=OneHotEncoder()
```

```python
encoded=encode_ohe.fit_transform(df[['species']]).toarray()
```

```python
encoded_df=pd.DataFrame(encoded,columns=encode_ohe.get_feature_names_out())
```
